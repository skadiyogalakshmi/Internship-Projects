```python
Subject = ['Maths','Science','English','Computer','Social']
df['Total'] = df[Subject].sum(axis=1)
df[['Student_Name','Total']].head(7)
```

```python
df['Average'] = df[Subject].mean(axis=1)
df[['Student_ID','Average']].tail(6)
```

```python
df.columns
```

Assigning Grades


```python
def grade(avg):
  if avg>=90:
    return 'A+'
  elif avg>=80:
    return 'A'
  elif avg>=70:
    return 'B'
  elif avg>=60:
    return 'C'
  elif avg>=50:
    return 'D'
  else:
    return 'F'
df['Grade'] = df['Average'].apply(grade)
df[['Student_Name','Average','Grade']].head(7)
```

Rank Students

```python
df["Rank"] = df["Average"].rank(
    ascending=False,
    method="dense"
).astype(int)

print(df[["Student_Name","Average","Rank"]].sort_values("Rank"))
```

Weak Students (Average < 40)

```python
weak_students = df[df["Average"] < 40]

print(weak_students)
```

Visualize Top 10 students (Bar chart)

```python
import matplotlib.pyplot as plt

# Optional: better figure size
plt.rcParams["figure.figsize"] = (8,5)
top10 = df.sort_values("Total", ascending=False).head(10)

plt.bar(top10["Student_Name"], top10["Total"])
plt.title("Top 10 Students")
plt.xlabel("Student Name")
plt.ylabel("Total Marks")
plt.xticks(rotation=45)

plt.show()
```

Subject-wise Average Marks (Bar Chart)

```python
subject_avg = df[Subject].mean()

plt.bar(subject_avg.index, subject_avg.values)

plt.title("Subject-wise Average Marks")
plt.xlabel("Subjects")
plt.ylabel("Average Marks")

plt.show()
```

Average Marks Distribution (Histogram)

```python
plt.hist(df["Average"], bins=10)

plt.title("Average Marks Distribution")
plt.xlabel("Average Marks")
plt.ylabel("Number of Students")

plt.show()
```

Pass vs Fail (Pie Chart)

```python
result_count = df["Pass_Fail"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(
    result_count.values,
    labels=result_count.index,
    autopct="%1.1f%%"
)

plt.title("Pass vs Fail Analysis")
plt.show()
```

Box Plot for Subject Performance

```python
plt.figure(figsize=(10,5))
plt.boxplot(
    [df["Maths"],
     df["Science"],
     df["English"],
     df["Computer"],
     df["Social"]],
    labels=Subject
)

plt.title("Subject-wise Marks Distribution")
plt.ylabel("Marks")
plt.show()
```

Subject Topper Marks Comparison(Line Chart)

```python
subjects = ["Maths","Science","English","Computer","Social"]

top_marks = [df[s].max() for s in subjects]

plt.figure(figsize=(8,5))
plt.plot(subjects,
         top_marks,
         marker='o')

plt.title("Subject Topper Marks")
plt.xlabel("Subjects")
plt.ylabel("Highest Marks")
plt.grid(True)
plt.show()
```
